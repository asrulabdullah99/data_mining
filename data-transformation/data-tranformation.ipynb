```python
import pandas as pd 
import numpy as np

kolom = {'col1' : [2,9,19],
        'col2' : [5,np.nan,17],
        'col3' : [3,9,np.nan],
        'col4' : [6,0,9],
        'col5' : [np.nan,7,np.nan]}

data = pd.DataFrame(kolom)
```

```python
data 
```

```python
#imputasi dengan mean
data.fillna(data.mean())
```

```python
#imputasi dengan median
data.fillna(data.median())
```

```python
import pandas as pd
import numpy as np 
from sklearn.impute import SimpleImputer

make = {'make':['Ford','Ford','Fiat','BMW','Fiat','BMW','Kia',np.nan,np.nan,'Fiat','Fiat']}
```

```python
data = pd.DataFrame(make)
data
```

```python
imp = SimpleImputer(strategy='most_frequent')
imp.fit_transform(data)
```
